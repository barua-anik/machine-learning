```python
"""
Logistic regression is the appropriate regression analysis
to conduct when the dependent variable is dichotomous (binary).
Although the name is regression, but it is a classification problem.
For example, we can get the answer Yes or No, Cancer or Not cancer.
It uses a sigmoid function, that can get the binary classification.
Same as other models, the logistic regression is a predictive analysis.

"""
```

```python
# Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the dataset

dataset = pd.read_csv('Social_Network_Ads.csv')
dataset.head()
```

```python
# Spliting the independent and the target variable 

X = dataset.iloc[ : , [2,3]].values.astype(float) # [2,3] and 2:4 is same
y = dataset.iloc[ : , 4 ].values.astype(float)
```

```python
# Spliting the dataset 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
# Feature scaling 

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
# Fiting logistic regression to the training set
from sklearn.linear_model import LogisticRegression # LogisticRegression class 
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
```

```python
# Predicting the test case result 

y_pred = classifier.predict(X_test)
```

```python
# Making the consusion matrix

from sklearn.metrics import confusion_matrix # confusion_matrix is a function
cm = confusion_matrix(y_test, y_pred)
```

```python
# Visualising the Training set results

from matplotlib.colors import ListedColormap
plt.rcParams["figure.figsize"] = (20,10)

X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)): # looping over all the data points
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
